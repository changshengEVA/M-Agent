```python
import requests
import json

url = "http://127.0.0.1:8777/v1/auth/login"

payload = json.dumps({
  "username": "test_1",
  "password": "testfirst"
})
headers = {
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)

assert response.status_code == 200
assert "access_token" in response.text
```
